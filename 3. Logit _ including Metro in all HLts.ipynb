```python
import numpy as np
import pandas as pd
df = pd.read_csv('data/hotels_merged.csv')
df=df.drop(columns = 'Unnamed: 0')
```

```python
df.sort_values("hotelName", inplace = True) 
```

```python
df.drop_duplicates(subset ="hotelName", 
                     keep = 'first', inplace = True)
```

```python
df.drop(df.columns[[10,11,12,13,14,15,16,17]],axis =1, inplace = True)
```

```python
df.drop(df[df['priceRange']=='none'].index, axis =0, inplace = True)

df['priceRange'].unique()

df['priceRange'].replace({'$':'1', '$$':'2','$$$':'3','$$$$':'4'}
                          , inplace = True)

df['priceRange'] = pd.to_numeric(df['priceRange'])
```

## Effect of variables on price

```python
import statsmodels.api as sm
model= sm.MNLogit(df['priceRange'], df[['hotelRate','totalRatings','totalPhotos','metro']])
answer= model.fit()
answer.summary()
```

### interpretation

> high Pseudo R-squ.:	0.1756

> Rate & total photo is positively correlated to the priceRange between 3-4
