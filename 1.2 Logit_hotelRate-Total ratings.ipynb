```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('data/hotels_merged.csv')
```

```python
df.sort_values("hotelName", inplace = True) 

df.drop_duplicates(subset ="hotelName", 
                     keep = 'first', inplace = True)

df.drop(df.columns[[10,11,12,13,14,15,16,17]],axis =1, inplace = True)
```

## data cleaning: Price 

```python
df.drop(df[df['priceRange']=='none'].index, axis =0, inplace = True)

df['priceRange'].unique()

df['priceRange'].replace({'$':'1', '$$':'2','$$$':'3','$$$$':'4'}
                          , inplace = True)

df['priceRange'] = pd.to_numeric(df['priceRange'])
```

# MN Logit - HR-#ratings

```python
model1= sm.MNLogit(df['hotelRate'], df['totalRatings'])
answer1= model1.fit()
answer1.summary()
```

## After dropping outliers

```python
model1_drop= sm.MNLogit(df_d1['hotelRate'], df_d1['totalRatings'])
answer1_drop= model1_drop.fit()
answer1_drop.summary()
```
